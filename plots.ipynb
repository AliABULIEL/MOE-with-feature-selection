```python
"""
=============================================================================
HC ROUTING VISUALIZATION - ENHANCED HD VERSION
=============================================================================
Publication-quality plots with high resolution and professional styling.

Features:
- 300 DPI for print quality
- Professional color palettes
- Clean typography
- Multiple output formats (PNG + PDF)
- Consistent styling across all plots
"""

import json
import os
from pathlib import Path
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
from matplotlib.patches import Patch
from matplotlib.lines import Line2D
import warnings
warnings.filterwarnings('ignore')

# =============================================================================
# CONFIGURATION - UPDATE THESE PATHS!
# =============================================================================
LOGS_DIR = Path('/Users/aliab/Downloads/logs')
VIZ_DIR = Path('/Users/aliab/Desktop/GitHub/MOE-with-feature-selection/plots')
VIZ_DIR.mkdir(parents=True, exist_ok=True)

# =============================================================================
# PROFESSIONAL PLOT STYLING
# =============================================================================
def set_publication_style():
    """Set publication-quality plot styling."""
    
    # Use a clean style
    plt.style.use('seaborn-v0_8-whitegrid')
    
    # High-quality settings
    plt.rcParams.update({
        # Figure
        'figure.figsize': (12, 8),
        'figure.dpi': 150,
        'figure.facecolor': 'white',
        'figure.edgecolor': 'white',
        'figure.autolayout': True,
        
        # Saving
        'savefig.dpi': 300,
        'savefig.format': 'png',
        'savefig.bbox': 'tight',
        'savefig.pad_inches': 0.1,
        'savefig.facecolor': 'white',
        'savefig.edgecolor': 'white',
        
        # Font
        'font.family': 'sans-serif',
        'font.sans-serif': ['Arial', 'DejaVu Sans', 'Helvetica', 'sans-serif'],
        'font.size': 12,
        'font.weight': 'normal',
        
        # Axes
        'axes.titlesize': 16,
        'axes.titleweight': 'bold',
        'axes.labelsize': 13,
        'axes.labelweight': 'medium',
        'axes.linewidth': 1.2,
        'axes.edgecolor': '#333333',
        'axes.facecolor': 'white',
        'axes.grid': True,
        'axes.spines.top': False,
        'axes.spines.right': False,
        
        # Grid
        'grid.alpha': 0.3,
        'grid.linewidth': 0.8,
        'grid.color': '#cccccc',
        
        # Legend
        'legend.fontsize': 11,
        'legend.frameon': True,
        'legend.framealpha': 0.95,
        'legend.edgecolor': '#cccccc',
        'legend.fancybox': True,
        
        # Ticks
        'xtick.labelsize': 11,
        'ytick.labelsize': 11,
        'xtick.major.width': 1.0,
        'ytick.major.width': 1.0,
        
        # Lines
        'lines.linewidth': 2.5,
        'lines.markersize': 8,
    })

# Professional color palettes
COLORS = {
    'primary': '#2563EB',      # Blue
    'secondary': '#10B981',    # Green
    'accent': '#F59E0B',       # Amber
    'danger': '#EF4444',       # Red
    'purple': '#8B5CF6',       # Purple
    'cyan': '#06B6D4',         # Cyan
    'pink': '#EC4899',         # Pink
    'gray': '#6B7280',         # Gray
}

# Color palette for multiple lines
LINE_COLORS = ['#2563EB', '#10B981', '#F59E0B', '#EF4444', '#8B5CF6', '#06B6D4']
BASELINE_COLOR = '#DC2626'  # Red for baseline reference

# =============================================================================
# HELPER FUNCTIONS
# =============================================================================

def save_plot(fig, name, viz_dir):
    """Save plot in multiple formats with high quality."""
    # PNG (high res)
    png_path = viz_dir / f'{name}.png'
    fig.savefig(png_path, dpi=300, bbox_inches='tight', facecolor='white', edgecolor='none')
    print(f"  ‚úÖ Saved: {png_path}")
    
    # PDF (vector)
    pdf_path = viz_dir / f'{name}.pdf'
    fig.savefig(pdf_path, format='pdf', bbox_inches='tight', facecolor='white', edgecolor='none')
    print(f"  ‚úÖ Saved: {pdf_path}")


def add_baseline_reference(ax, y=8, label='Baseline K=8'):
    """Add baseline reference line to plot."""
    ax.axhline(y=y, color=BASELINE_COLOR, linestyle='--', linewidth=2, alpha=0.8, label=label, zorder=1)


def format_config_name(name):
    """Format config name for display."""
    name = name.replace('experts_', '')
    name = name.replace('_', ' ')
    name = name.replace('topk baseline', 'TopK')
    name = name.replace('hc ', 'HC ')
    return name

# =============================================================================
# LOAD DATA
# =============================================================================

print("=" * 70)
print("HC ROUTING VISUALIZATION - ENHANCED HD VERSION")
print("=" * 70)

# Set style
set_publication_style()

# Load results
print(f"\nüìÅ Loading results from: {LOGS_DIR}")
results = []
json_files = list(LOGS_DIR.glob('*.json'))
print(f"   Found {len(json_files)} JSON files")

for json_file in json_files:
    with open(json_file, 'r') as f:
        results.append(json.load(f))

df = pd.DataFrame(results)
print(f"   ‚úÖ Loaded {len(df)} experiment results")
print(f"   Datasets: {df['dataset'].unique().tolist()}")
print(f"   Routing types: {df['routing_type'].unique().tolist()}")

# Separate HC and baseline
hc_df = df[df['routing_type'] == 'hc'].copy()
baseline_df = df[df['routing_type'] == 'topk'].copy()

# =============================================================================
# PLOT 1: AVERAGE EXPERTS BY CONFIGURATION (Enhanced)
# =============================================================================
print("\n" + "=" * 70)
print("PLOT 1: Average Experts by Configuration")
print("=" * 70)

fig, axes = plt.subplots(1, 3, figsize=(20, 8))
fig.suptitle('Average Expert Selection by Configuration', fontsize=20, fontweight='bold', y=1.02)

for idx, dataset in enumerate(['wikitext', 'lambada', 'hellaswag']):
    ax = axes[idx]
    subset = df[df['dataset'] == dataset].sort_values('avg_experts')
    
    if len(subset) == 0:
        ax.text(0.5, 0.5, f'No data for {dataset}', ha='center', va='center', fontsize=14)
        ax.set_title(dataset.upper(), fontsize=16, fontweight='bold')
        continue
    
    # Create color array
    colors = [COLORS['secondary'] if rt == 'topk' else COLORS['primary'] for rt in subset['routing_type']]
    
    # Create horizontal bar chart
    y_pos = np.arange(len(subset))
    bars = ax.barh(y_pos, subset['avg_experts'], color=colors, edgecolor='white', linewidth=0.5, height=0.7)
    
    # Styling
    ax.set_yticks(y_pos)
    ax.set_yticklabels([format_config_name(n) for n in subset['config_name']], fontsize=9)
    ax.set_xlabel('Average Experts per Token', fontsize=12, fontweight='medium')
    ax.set_title(f'{dataset.upper()}', fontsize=16, fontweight='bold')
    
    # Baseline reference line
    ax.axvline(x=8, color=BASELINE_COLOR, linestyle='--', linewidth=2.5, alpha=0.8, zorder=10)
    
    # Value labels
    for i, (bar, val) in enumerate(zip(bars, subset['avg_experts'])):
        ax.text(val + 0.5, i, f'{val:.1f}', va='center', fontsize=9, fontweight='medium')
    
    # Set x-axis limit
    ax.set_xlim(0, max(subset['avg_experts']) * 1.15)
    ax.set_ylim(-0.5, len(subset) - 0.5)

# Create legend
legend_elements = [
    Patch(facecolor=COLORS['secondary'], edgecolor='white', label='TopK Baseline'),
    Patch(facecolor=COLORS['primary'], edgecolor='white', label='HC Routing'),
    Line2D([0], [0], color=BASELINE_COLOR, linestyle='--', linewidth=2.5, label='K=8 Reference')
]
fig.legend(handles=legend_elements, loc='upper center', bbox_to_anchor=(0.5, 0.98), 
           ncol=3, fontsize=12, frameon=True)

plt.tight_layout()
save_plot(fig, 'plot1_experts_by_config', VIZ_DIR)
plt.show()

# =============================================================================
# PLOT 2: BETA EFFECT ON EXPERT SELECTION (Enhanced)
# =============================================================================
print("\n" + "=" * 70)
print("PLOT 2: Beta Effect on Expert Selection")
print("=" * 70)

fig, axes = plt.subplots(2, 2, figsize=(16, 14))
fig.suptitle('Effect of Beta Parameter (Œ≤) on HC Routing', fontsize=20, fontweight='bold', y=1.01)

# Plot 2a: Avg Experts vs Beta by max_k
ax1 = axes[0, 0]
for i, max_k in enumerate(sorted(hc_df['k_or_max_k'].unique())):
    subset = hc_df[(hc_df['k_or_max_k'] == max_k) & (hc_df['dataset'] == 'wikitext')]
    if len(subset) > 0:
        subset = subset.sort_values('beta')
        ax1.plot(subset['beta'], subset['avg_experts'], 
                marker='o', linewidth=2.5, markersize=10, 
                color=LINE_COLORS[i % len(LINE_COLORS)],
                label=f'max_k = {int(max_k)}')

add_baseline_reference(ax1)
ax1.set_xlabel('Beta (Œ≤)', fontsize=13, fontweight='medium')
ax1.set_ylabel('Average Experts', fontsize=13, fontweight='medium')
ax1.set_title('Average Experts vs Beta (WikiText)', fontsize=14, fontweight='bold')
ax1.legend(loc='upper left', fontsize=11)
ax1.set_xlim(0.25, 0.65)

# Plot 2b: Average experts by beta (all configs)
ax2 = axes[0, 1]
beta_means = hc_df.groupby('beta')['avg_experts'].mean()
colors = plt.cm.viridis(np.linspace(0.2, 0.8, len(beta_means)))
bars = ax2.bar(range(len(beta_means)), beta_means.values, color=colors, 
               edgecolor='white', linewidth=1.5, width=0.7)
ax2.set_xticks(range(len(beta_means)))
ax2.set_xticklabels([f'Œ≤ = {b:.2f}' for b in beta_means.index], fontsize=11)
ax2.set_ylabel('Average Experts (Mean)', fontsize=13, fontweight='medium')
ax2.set_title('Mean Expert Count by Beta', fontsize=14, fontweight='bold')
ax2.axhline(y=8, color=BASELINE_COLOR, linestyle='--', linewidth=2.5, alpha=0.8)

# Value labels
for bar, val in zip(bars, beta_means.values):
    ax2.text(bar.get_x() + bar.get_width()/2, val + 0.5, f'{val:.1f}', 
            ha='center', fontsize=11, fontweight='bold')

# Plot 2c: Std Experts vs Beta
ax3 = axes[1, 0]
for i, max_k in enumerate(sorted(hc_df['k_or_max_k'].unique())):
    subset = hc_df[(hc_df['k_or_max_k'] == max_k) & (hc_df['dataset'] == 'wikitext')]
    if len(subset) > 0:
        subset = subset.sort_values('beta')
        ax3.plot(subset['beta'], subset['std_experts'], 
                marker='s', linewidth=2.5, markersize=10, 
                color=LINE_COLORS[i % len(LINE_COLORS)],
                label=f'max_k = {int(max_k)}')

ax3.set_xlabel('Beta (Œ≤)', fontsize=13, fontweight='medium')
ax3.set_ylabel('Std Experts', fontsize=13, fontweight='medium')
ax3.set_title('Expert Variability vs Beta', fontsize=14, fontweight='bold')
ax3.legend(loc='upper left', fontsize=11)
ax3.set_xlim(0.25, 0.65)

# Plot 2d: Adaptive Range vs Beta
ax4 = axes[1, 1]
if 'adaptive_range' in hc_df.columns:
    for i, max_k in enumerate(sorted(hc_df['k_or_max_k'].unique())):
        subset = hc_df[(hc_df['k_or_max_k'] == max_k) & (hc_df['dataset'] == 'wikitext')]
        if len(subset) > 0:
            subset = subset.sort_values('beta')
            ax4.plot(subset['beta'], subset['adaptive_range'], 
                    marker='^', linewidth=2.5, markersize=10, 
                    color=LINE_COLORS[i % len(LINE_COLORS)],
                    label=f'max_k = {int(max_k)}')

ax4.set_xlabel('Beta (Œ≤)', fontsize=13, fontweight='medium')
ax4.set_ylabel('Adaptive Range (max - min)', fontsize=13, fontweight='medium')
ax4.set_title('Adaptive Range vs Beta', fontsize=14, fontweight='bold')
ax4.legend(loc='upper left', fontsize=11)
ax4.set_xlim(0.25, 0.65)

plt.tight_layout()
save_plot(fig, 'plot2_beta_effect', VIZ_DIR)
plt.show()

# =============================================================================
# PLOT 3: EXPERT SELECTION HEATMAP (Enhanced)
# =============================================================================
print("\n" + "=" * 70)
print("PLOT 3: Expert Selection Heatmap")
print("=" * 70)

wiki_hc = hc_df[hc_df['dataset'] == 'wikitext']

if len(wiki_hc) > 0:
    pivot = wiki_hc.pivot_table(values='avg_experts', index='beta', columns='k_or_max_k')
    
    fig, ax = plt.subplots(figsize=(12, 9))
    
    # Create heatmap
    sns.heatmap(pivot, annot=True, fmt='.1f', cmap='RdYlBu_r', ax=ax,
                cbar_kws={'label': 'Average Experts', 'shrink': 0.8},
                annot_kws={'size': 14, 'weight': 'bold'},
                linewidths=2, linecolor='white',
                square=True)
    
    # Styling
    ax.set_xlabel('max_k', fontsize=14, fontweight='bold')
    ax.set_ylabel('Beta (Œ≤)', fontsize=14, fontweight='bold')
    ax.set_title('Average Experts Heatmap\n(WikiText Dataset)', fontsize=18, fontweight='bold', pad=20)
    
    # Format tick labels
    ax.set_xticklabels([f'{int(x)}' for x in pivot.columns], fontsize=12)
    ax.set_yticklabels([f'{y:.2f}' for y in pivot.index], fontsize=12, rotation=0)
    
    plt.tight_layout()
    save_plot(fig, 'plot3_experts_heatmap', VIZ_DIR)
    plt.show()

# =============================================================================
# PLOT 4: QUALITY VS EFFICIENCY (Enhanced)
# =============================================================================
print("\n" + "=" * 70)
print("PLOT 4: Quality vs Efficiency Trade-off")
print("=" * 70)

fig, axes = plt.subplots(1, 3, figsize=(20, 7))
fig.suptitle('Quality vs Efficiency Trade-off', fontsize=20, fontweight='bold', y=1.02)

datasets_info = [
    ('wikitext', 'perplexity', 'Perplexity (‚Üì better)', False),
    ('lambada', 'lambada_accuracy', 'LAMBADA Accuracy (‚Üë better)', True),
    ('hellaswag', 'hellaswag_accuracy', 'HellaSwag Accuracy (‚Üë better)', True)
]

for idx, (dataset, metric, ylabel, higher_better) in enumerate(datasets_info):
    ax = axes[idx]
    subset = df[df['dataset'] == dataset]
    
    if metric not in subset.columns or subset[metric].isna().all():
        ax.text(0.5, 0.5, f'No {metric} data', ha='center', va='center', 
                transform=ax.transAxes, fontsize=14)
        ax.set_title(dataset.upper(), fontsize=16, fontweight='bold')
        continue
    
    baseline = subset[subset['routing_type'] == 'topk']
    hc = subset[subset['routing_type'] == 'hc']
    
    # Plot baseline
    if len(baseline) > 0 and baseline[metric].notna().any():
        ax.scatter(baseline['avg_experts'], baseline[metric], 
                  s=200, marker='s', c=COLORS['secondary'], 
                  label='TopK Baseline', zorder=5, edgecolor='white', linewidth=2)
        
        # Add labels for baseline points
        for _, row in baseline.iterrows():
            if pd.notna(row[metric]):
                ax.annotate(f"K={int(row['k_or_max_k'])}", 
                           (row['avg_experts'], row[metric]),
                           textcoords="offset points", xytext=(5, 5),
                           fontsize=9, fontweight='medium')
    
    # Plot HC
    if len(hc) > 0 and hc[metric].notna().any():
        scatter = ax.scatter(hc['avg_experts'], hc[metric], 
                            s=150, c=hc['beta'], cmap='plasma', 
                            label='HC Routing', alpha=0.85, edgecolor='white', linewidth=1.5)
        cbar = plt.colorbar(scatter, ax=ax, shrink=0.8, pad=0.02)
        cbar.set_label('Beta (Œ≤)', fontsize=11, fontweight='medium')
    
    ax.set_xlabel('Average Experts', fontsize=13, fontweight='medium')
    ax.set_ylabel(ylabel, fontsize=13, fontweight='medium')
    ax.set_title(f'{dataset.upper()}', fontsize=16, fontweight='bold')
    ax.legend(loc='best', fontsize=10)
    
    # Add baseline reference
    ax.axvline(x=8, color=BASELINE_COLOR, linestyle='--', linewidth=2, alpha=0.6)

plt.tight_layout()
save_plot(fig, 'plot4_quality_vs_efficiency', VIZ_DIR)
plt.show()

# =============================================================================
# PLOT 5: FLOOR/CEILING ANALYSIS (Enhanced)
# =============================================================================
print("\n" + "=" * 70)
print("PLOT 5: Floor/Ceiling Hit Rate Analysis")
print("=" * 70)

if 'ceiling_hit_rate' in hc_df.columns and 'floor_hit_rate' in hc_df.columns:
    wiki_hc = hc_df[hc_df['dataset'] == 'wikitext'].sort_values(['k_or_max_k', 'beta'])
    
    if len(wiki_hc) > 0:
        fig, axes = plt.subplots(1, 3, figsize=(20, 7))
        fig.suptitle('Floor/Ceiling Hit Rate Analysis (WikiText)', fontsize=20, fontweight='bold', y=1.02)
        
        # Plot 5a: Stacked bar
        ax1 = axes[0]
        configs = wiki_hc['config_name'].values
        x = np.arange(len(configs))
        width = 0.7
        
        floor_vals = wiki_hc['floor_hit_rate'].values
        mid_vals = wiki_hc['mid_range_rate'].values
        ceiling_vals = wiki_hc['ceiling_hit_rate'].values
        
        p1 = ax1.bar(x, floor_vals, width, label='Floor (min_k)', color=COLORS['danger'], edgecolor='white')
        p2 = ax1.bar(x, mid_vals, width, bottom=floor_vals, label='Mid Range', color=COLORS['primary'], edgecolor='white')
        p3 = ax1.bar(x, ceiling_vals, width, bottom=floor_vals+mid_vals, label='Ceiling (max_k)', color=COLORS['secondary'], edgecolor='white')
        
        ax1.set_xlabel('Configuration', fontsize=13, fontweight='medium')
        ax1.set_ylabel('Percentage (%)', fontsize=13, fontweight='medium')
        ax1.set_title('Token Distribution by Expert Selection', fontsize=14, fontweight='bold')
        ax1.set_xticks(x)
        ax1.set_xticklabels([format_config_name(c) for c in configs], rotation=45, ha='right', fontsize=9)
        ax1.legend(loc='upper right', fontsize=10)
        ax1.set_ylim(0, 105)
        
        # Plot 5b: Mid-range rate by beta
        ax2 = axes[1]
        for i, max_k in enumerate(sorted(wiki_hc['k_or_max_k'].unique())):
            subset = wiki_hc[wiki_hc['k_or_max_k'] == max_k].sort_values('beta')
            ax2.plot(subset['beta'], subset['mid_range_rate'], 
                    marker='o', linewidth=2.5, markersize=10, 
                    color=LINE_COLORS[i % len(LINE_COLORS)],
                    label=f'max_k = {int(max_k)}')
        
        ax2.set_xlabel('Beta (Œ≤)', fontsize=13, fontweight='medium')
        ax2.set_ylabel('Mid-Range Rate (%)', fontsize=13, fontweight='medium')
        ax2.set_title('Adaptive Selection Rate\n(Higher = More Dynamic)', fontsize=14, fontweight='bold')
        ax2.legend(loc='best', fontsize=10)
        ax2.set_xlim(0.25, 0.65)
        
        # Plot 5c: Floor hit rate
        ax3 = axes[2]
        for i, max_k in enumerate(sorted(wiki_hc['k_or_max_k'].unique())):
            subset = wiki_hc[wiki_hc['k_or_max_k'] == max_k].sort_values('beta')
            ax3.plot(subset['beta'], subset['floor_hit_rate'], 
                    marker='s', linewidth=2.5, markersize=10, 
                    color=LINE_COLORS[i % len(LINE_COLORS)],
                    label=f'max_k = {int(max_k)}')
        
        ax3.set_xlabel('Beta (Œ≤)', fontsize=13, fontweight='medium')
        ax3.set_ylabel('Floor Hit Rate (%)', fontsize=13, fontweight='medium')
        ax3.set_title('Minimum Expert Usage\n(Lower = Better Adaptivity)', fontsize=14, fontweight='bold')
        ax3.legend(loc='best', fontsize=10)
        ax3.set_xlim(0.25, 0.65)
        
        plt.tight_layout()
        save_plot(fig, 'plot5_floor_ceiling_analysis', VIZ_DIR)
        plt.show()

# =============================================================================
# PLOT 6: SUMMARY COMPARISON (Enhanced)
# =============================================================================
print("\n" + "=" * 70)
print("PLOT 6: Summary Comparison")
print("=" * 70)

fig, axes = plt.subplots(2, 2, figsize=(16, 14))
fig.suptitle('HC Routing Experiment Summary', fontsize=20, fontweight='bold', y=1.01)

# Plot 6a: Box plot of experts by routing type
ax1 = axes[0, 0]
plot_data = df[df['dataset'].isin(['wikitext', 'lambada', 'hellaswag'])].copy()
if len(plot_data) > 0:
    palette = {'topk': COLORS['secondary'], 'hc': COLORS['primary']}
    sns.boxplot(data=plot_data, x='dataset', y='avg_experts', hue='routing_type',
               ax=ax1, palette=palette, width=0.6)
    ax1.axhline(y=8, color=BASELINE_COLOR, linestyle='--', linewidth=2.5, alpha=0.8)
    ax1.set_title('Expert Distribution by Routing Type', fontsize=14, fontweight='bold')
    ax1.set_xlabel('Dataset', fontsize=13, fontweight='medium')
    ax1.set_ylabel('Average Experts', fontsize=13, fontweight='medium')
    ax1.legend(title='Routing Type', fontsize=10)

# Plot 6b: Violin plot of expert variability
ax2 = axes[0, 1]
if 'std_experts' in hc_df.columns and len(hc_df) > 0:
    palette = plt.cm.viridis(np.linspace(0.2, 0.8, len(hc_df['beta'].unique())))
    sns.violinplot(data=hc_df, x='beta', y='std_experts', ax=ax2, palette='viridis')
    ax2.set_title('Expert Variability by Beta', fontsize=14, fontweight='bold')
    ax2.set_xlabel('Beta (Œ≤)', fontsize=13, fontweight='medium')
    ax2.set_ylabel('Std Experts', fontsize=13, fontweight='medium')

# Plot 6c: Bar chart comparing means
ax3 = axes[1, 0]
summary = df.groupby(['routing_type', 'dataset'])['avg_experts'].mean().reset_index()
if len(summary) > 0:
    pivot = summary.pivot(index='dataset', columns='routing_type', values='avg_experts')
    pivot.plot(kind='bar', ax=ax3, width=0.7, color=[COLORS['primary'], COLORS['secondary']], edgecolor='white')
    ax3.axhline(y=8, color=BASELINE_COLOR, linestyle='--', linewidth=2.5, alpha=0.8)
    ax3.set_title('Mean Experts by Routing Type', fontsize=14, fontweight='bold')
    ax3.set_xlabel('Dataset', fontsize=13, fontweight='medium')
    ax3.set_ylabel('Average Experts', fontsize=13, fontweight='medium')
    ax3.legend(title='Routing Type', fontsize=10)
    ax3.tick_params(axis='x', rotation=0)

# Plot 6d: Scatter of HC configs
ax4 = axes[1, 1]
if len(hc_df) > 0:
    scatter = ax4.scatter(hc_df['k_or_max_k'], hc_df['avg_experts'], 
                         c=hc_df['beta'], s=150, cmap='plasma', 
                         alpha=0.8, edgecolor='white', linewidth=1.5)
    cbar = plt.colorbar(scatter, ax=ax4, shrink=0.8)
    cbar.set_label('Beta (Œ≤)', fontsize=11, fontweight='medium')
    ax4.axhline(y=8, color=BASELINE_COLOR, linestyle='--', linewidth=2.5, alpha=0.8)
    ax4.set_xlabel('max_k', fontsize=13, fontweight='medium')
    ax4.set_ylabel('Average Experts', fontsize=13, fontweight='medium')
    ax4.set_title('HC Configurations Overview', fontsize=14, fontweight='bold')

plt.tight_layout()
save_plot(fig, 'plot6_summary_comparison', VIZ_DIR)
plt.show()

# =============================================================================
# PLOT 7: KEY FINDINGS VISUALIZATION (New!)
# =============================================================================
print("\n" + "=" * 70)
print("PLOT 7: Key Findings")
print("=" * 70)

fig, axes = plt.subplots(1, 2, figsize=(16, 7))
fig.suptitle('HC Routing: Key Findings', fontsize=20, fontweight='bold', y=1.02)

# Plot 7a: Beta Effect Summary
ax1 = axes[0]
wiki_hc_summary = hc_df[hc_df['dataset'] == 'wikitext'].groupby('beta').agg({
    'avg_experts': 'mean',
    'std_experts': 'mean'
}).reset_index()

if len(wiki_hc_summary) > 0:
    x = np.arange(len(wiki_hc_summary))
    width = 0.35
    
    bars1 = ax1.bar(x - width/2, wiki_hc_summary['avg_experts'], width, 
                   label='Avg Experts', color=COLORS['primary'], edgecolor='white')
    bars2 = ax1.bar(x + width/2, wiki_hc_summary['std_experts'], width, 
                   label='Std Experts', color=COLORS['accent'], edgecolor='white')
    
    ax1.axhline(y=8, color=BASELINE_COLOR, linestyle='--', linewidth=2.5, alpha=0.8, label='Baseline K=8')
    ax1.set_xlabel('Beta (Œ≤)', fontsize=13, fontweight='medium')
    ax1.set_ylabel('Expert Count', fontsize=13, fontweight='medium')
    ax1.set_title('Beta Controls Expert Selection', fontsize=14, fontweight='bold')
    ax1.set_xticks(x)
    ax1.set_xticklabels([f'{b:.2f}' for b in wiki_hc_summary['beta']], fontsize=11)
    ax1.legend(loc='upper left', fontsize=10)
    
    # Value labels
    for bar in bars1:
        height = bar.get_height()
        ax1.text(bar.get_x() + bar.get_width()/2, height + 0.3, f'{height:.1f}',
                ha='center', fontsize=10, fontweight='bold')

# Plot 7b: Saturation Analysis
ax2 = axes[1]
saturation_data = hc_df[hc_df['dataset'] == 'wikitext'].groupby('k_or_max_k')['avg_experts'].mean()

if len(saturation_data) > 0:
    ax2.plot(saturation_data.index, saturation_data.values, 
            marker='o', linewidth=3, markersize=12, color=COLORS['primary'])
    ax2.axhline(y=8, color=BASELINE_COLOR, linestyle='--', linewidth=2.5, alpha=0.8, label='Baseline K=8')
    ax2.fill_between(saturation_data.index, saturation_data.values, alpha=0.3, color=COLORS['primary'])
    
    ax2.set_xlabel('max_k', fontsize=13, fontweight='medium')
    ax2.set_ylabel('Average Experts', fontsize=13, fontweight='medium')
    ax2.set_title('HC Saturation by max_k', fontsize=14, fontweight='bold')
    ax2.legend(loc='upper left', fontsize=10)
    
    # Value labels
    for x, y in zip(saturation_data.index, saturation_data.values):
        ax2.annotate(f'{y:.1f}', (x, y), textcoords="offset points", 
                    xytext=(0, 10), ha='center', fontsize=11, fontweight='bold')

plt.tight_layout()
save_plot(fig, 'plot7_key_findings', VIZ_DIR)
plt.show()

# =============================================================================
# SAVE SUMMARY TABLE
# =============================================================================
print("\n" + "=" * 70)
print("SAVING RESULTS SUMMARY")
print("=" * 70)

# Create summary
for dataset in df['dataset'].unique():
    subset = df[df['dataset'] == dataset].sort_values('avg_experts')
    print(f"\nüìä {dataset.upper()}")
    print("-" * 60)
    
    cols = ['config_name', 'routing_type', 'avg_experts', 'std_experts']
    if 'perplexity' in subset.columns and subset['perplexity'].notna().any():
        cols.append('perplexity')
    if 'lambada_accuracy' in subset.columns and subset['lambada_accuracy'].notna().any():
        cols.append('lambada_accuracy')
    if 'hellaswag_accuracy' in subset.columns and subset['hellaswag_accuracy'].notna().any():
        cols.append('hellaswag_accuracy')
    
    display_cols = [c for c in cols if c in subset.columns]
    print(subset[display_cols].to_string(index=False))

# Save CSV
csv_path = VIZ_DIR / 'all_results_summary.csv'
df.to_csv(csv_path, index=False)
print(f"\n‚úÖ Saved CSV: {csv_path}")

print("\n" + "=" * 70)
print("‚úÖ ALL HD VISUALIZATIONS COMPLETE!")
print(f"üìÅ All plots saved to: {VIZ_DIR}")
print("   ‚Ä¢ PNG format (300 DPI)")
print("   ‚Ä¢ PDF format (vector)")
print("=" * 70)
```
